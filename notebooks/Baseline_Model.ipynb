# Basic Model

I implemented a basic model in order to just check that everything works since the multi-label prediction seems quite mystical. I train on a portion of the data and I predict a single point and check if the predictions seems reasonable (it does for most).

```python
import pandas as pd
import numpy as np
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
```

```python
#Load the previously transformed data.
NUM_FEATURES = 5000
NUM_CLASSES = 3993
features = pd.read_csv("../data/expanded/train_features.csv", names=range(NUM_FEATURES))
labels = pd.read_csv("../data/expanded/train_labels.csv", names=range(NUM_CLASSES))
```

# Train basic model

```python
#Training takes around 30min on my machine. 5000 classes  = 5000 seperate models in one-vs-rest scheme so not too efficient.
model = OneVsRestClassifier(LinearSVC(random_state=0)).fit(features[102:10000],labels[102:10000])
```

# Predict single point

```python
label_pred = model.predict(features.iloc[2,:].values.reshape(1, -1))
np.nonzero(label_pred)
```
